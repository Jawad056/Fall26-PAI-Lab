```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread("pyramid.jpg")

if image is None:
    raise FileNotFoundError("Image not found")

image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image_rgb)
plt.axis("off")
plt.show()
```
