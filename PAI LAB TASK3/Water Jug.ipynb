```python
JUG1_CAPACITY = 4
JUG2_CAPACITY = 3
GOAL = 2

visited = set()

def dfs(jug1, jug2, path):
    print(f"Current State: ({jug1}, {jug2})")

    if jug1 == GOAL or jug2 == GOAL:
        print("\nGoal Reached!")
        print("Path:")
        for step in path:
            print(step)
        return True

    if (jug1, jug2) in visited:
        return False

    visited.add((jug1, jug2))

    if dfs(JUG1_CAPACITY, jug2, path + [f"Fill Jug1  ({JUG1_CAPACITY}, {jug2})"]):
        return True


    if dfs(jug1, JUG2_CAPACITY, path + [f"Fill Jug2  ({jug1}, {JUG2_CAPACITY})"]):
        return True

    if dfs(0, jug2, path + [f"Empty Jug1  (0, {jug2})"]):
        return True

    if dfs(jug1, 0, path + [f"Empty Jug2 ({jug1}, 0)"]):
        return True

    transfer = min(jug1, JUG2_CAPACITY - jug2)
    new_jug1 = jug1 - transfer
    new_jug2 = jug2 + transfer
    if dfs(new_jug1, new_jug2, path + [f"Pour Jug1  Jug2  ({new_jug1}, {new_jug2})"]):
        return True

    transfer = min(jug2, JUG1_CAPACITY - jug1)
    new_jug1 = jug1 + transfer
    new_jug2 = jug2 - transfer
    if dfs(new_jug1, new_jug2, path + [f"Pour Jug2  Jug1  ({new_jug1}, {new_jug2})"]):
        return True

    return False


dfs(0, 0, [])
```
